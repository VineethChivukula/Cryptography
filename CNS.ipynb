```python
#VINNY Ceaser Cipher
#Encryption and Decryption
def caesar_cipher(text, shift, ch):
    ans = ""
    for char in text:
        char=char.lower()
        if char.isalpha() and ch==1:
            tempchar = chr((ord(char) - 97 + shift) % 26 + 97)
        elif char.isalpha() and ch==2:
            tempchar = chr((ord(char) - 97 - shift) % 26 + 97)
        else:
            tempchar = char
        ans += tempchar
    return ans
#Choose either Encrption or Decryption
print("List of operations:")
print("1.Encryption")
print("2.Decryption")
print("3.Exit")
while(True):
    ch=int(input("\nEnter the choice:"))
    if(ch==1):
        text=input("Plain Text:")
        shift=int(input("Shift:"))
        print("Cypher Text:",caesar_cipher(text,shift,ch))
    if(ch==2):
        text=input("Cipher Text:")
        shift=int(input("Shift:"))
        print("Plain Text:",caesar_cipher(text,shift,ch))
    else:
      break
```

```python
#Hill Cipher
import numpy as np
def create_matrix_from(key):
    return [[ord(key[3*i+j]) % 65 for j in range(3)] for i in range(3)]
def encrypt(P, K):
    return [(sum(K[i][j] * P[j] for j in range(3)) % 26) for i in range(3)]
def Hill(message, K):
    cipher_text = []
    for i in range(0, len(message), 3):
        P = [(ord(message[i+j]) - 65) % 26 for j in range(3)]
        C = encrypt(P, K)
        cipher_text.extend(chr(c + 65) for c in C)
    return ''.join(cipher_text)
def MatrixInverse(K):
    det = int(np.linalg.det(K))
    det_multiplicative_inverse = pow(det, -1, 26)
    K_inv = [[0] * 3 for i in range(3)]
    for i in range(3):
        for j in range(3):
            Dji = K
            Dji = np.delete(Dji, (j), axis=0)
            Dji = np.delete(Dji, (i), axis=1)
            det = Dji[0][0]*Dji[1][1] - Dji[0][1]*Dji[1][0]
            K_inv[i][j] = (det_multiplicative_inverse * pow(-1,i+j) * det) % 26
    return K_inv
message = input("Enter the plain text (length should be multiples of 3): ").upper()
key = input("Enter the key of length 9(3x3 matrix): ")
K = create_matrix_from(key)
print('K-Matrix: ',K)
cipher_text = Hill(message, K)
print ('Cipher text: ', cipher_text)
K_inv = MatrixInverse(K)
print("k-inv matrix: ",K_inv)
plain_text = Hill(cipher_text, K_inv)
print ('Plain text: ', plain_text)
```

```python
#Play Fair Cipher
key = input("Enter key: ").replace(" ", "").upper()
result = []
for c in key:
    if c not in result:
        if c == 'J':
            result.append('I')
        else:
            result.append(c)
result += [chr(i) if i != 73 else 'I' for i in range(65, 91) if chr(i) not in result and i != 74]
my_matrix = [[result[k] for k in range(i*5, i*5+5)] for i in range(5)]
def locindex(c):
    if c == 'J':
        c = 'I'
    for i, j in enumerate(my_matrix):
        if c in j:
            return [i, j.index(c)]
def process_text(text):
    text = text.upper().replace(" ", "")
    text = ''.join([text[i]+'X' if i<len(text)-1 and text[i]==text[i+1] else text[i] for i in range(len(text))])
    text += 'X' if len(text)%2 else ''
    return text
def do_work(msg, enc=True):
    output = []
    for i in range(0, len(msg), 2):
        loc1 = locindex(msg[i])
        loc2 = locindex(msg[i+1])
        if loc1[1] == loc2[1]:
            output.append("{}{}".format(my_matrix[(loc1[0]+(1 if enc else -1))%5][loc1[1]], my_matrix[(loc2[0]+(1 if enc else -1))%5][loc2[1]]))
        elif loc1[0] == loc2[0]:
            output.append("{}{}".format(my_matrix[loc1[0]][(loc1[1]+(1 if enc else -1))%5], my_matrix[loc2[0]][(loc2[1]+(1 if enc else -1))%5]))
        else:
            output.append("{}{}".format(my_matrix[loc1[0]][loc2[1]], my_matrix[loc2[0]][loc1[1]]))
    return ' '.join(output)
print("List of Operations:")
print("1.Encryption")
print("2.Decryption")
print("3.Exit")
while True:
    choice = input("\nEnter your choice: ")
    if choice == '1':
        msg = input("Plain Text: ")
        msg = process_text(msg)
        print("Cipher Text: {}".format(do_work(msg)))
    elif choice == '2':
        msg = input("Cipher Text: ")
        print("Plain Text: {}".format(do_work(msg, False)))
    elif choice == '3':
        break
```

```python
#Columnar Cipher
#Encryption and Decryption
import math
def columnar_cipher(msg, key, ch):
    key_order = sorted(range(len(key)), key=lambda i: key[i])
    n_cols = len(key)
    n_rows = int(math.ceil(len(msg) / n_cols))
    if ch==1:
        msg += '_' * (n_rows * n_cols - len(msg))
        msg_matrix = [msg[i:i+n_cols] for i in range(0, len(msg), n_cols)]
        cipher = ''.join([msg_matrix[i][j] for j in key_order for i in range(n_rows)])
        return cipher
    else:
        cipher_matrix = [[None]*n_cols for _ in range(n_rows)]
        p=0
        for j in key_order:
            for i in range(n_rows):
                if p < len(msg):
                    cipher_matrix[i][j] = msg[p]
                    p += 1
        plain = ''.join([cipher_matrix[i][j] for i in range(n_rows) for j in range(n_cols)])
        return plain.rstrip('_')
#Choose either Encrption or Decryption
print("List of operations:")
print("1.Encryption")
print("2.Decryption")
print("3.Exit")
while(True):
    ch=int(input("\nEnter the choice:"))
    if(ch==1):
        text=input("Plain Text:")
        key=input("key:")
        print("Cipher Text:",columnar_cipher(text,key,ch))
    if(ch==2):
        text=input("Cipher Text:")
        key=input("key:")
        print("Plain Text:",columnar_cipher(text,key,ch))
    else:
        break
```

```python
#DES
#pip install pyDes
import pyDes
# Encryption and Decryption
def DES(text, key, ch):
    if ch == 1:
        return pyDes.des(key, pyDes.ECB, pad=None, padmode=pyDes.PAD_PKCS5).encrypt(text).hex()
    if ch == 2:
        return pyDes.des(key, pyDes.ECB, pad=None, padmode=pyDes.PAD_PKCS5).decrypt(bytes.fromhex(text)).decode()
# Choose either Encryption or Decryption
print("List of operations:")
print("1.Encryption")
print("2.Decryption")
print("3.Exit")
while True:
    ch = int(input("\nEnter the choice:"))
    if ch == 1:
        text = input("Enter the message to encrypt: ")
        key = input("Enter the key for encryption: ")
        print("Cipher Text:", DES(text, key, ch))
    if ch == 2:
        text = input("Enter the message to decrypt: ")
        key = input("Enter the key for decryption: ")
        print("Plain Text:", DES(text, key, ch))
    else:
        break
```

```python
#AES
#pip install pycryptodome
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
#Encryption and Decryption
def AES_cipher(text, key, ch):
    if ch==1:
        return AES.new(key.encode('utf-8'), AES.MODE_ECB).encrypt(pad(text.encode('utf-8'), AES.block_size)).hex().upper()
    if ch==2:
        return unpad(AES.new(key.encode('utf-8'), AES.MODE_ECB).decrypt(bytes.fromhex(text)), AES.block_size).decode()
#Choose either Encrption or Decryption
print("List of operations:")
print("1.Encryption")
print("2.Decryption")
print("3.Exit")
while(True):
    ch=int(input("\nEnter the choice:"))
    if(ch==1):
        text = input("Enter the message to encrypt: ")
        key = input("Enter the key for encryption: ")
        print("Cipher Text:", AES_cipher(text, key, ch))
    if(ch==2):
        text = input("Enter the message to decrypt: ")
        key = input("Enter the key for decryption: ")
        print("Plain Text:", AES_cipher(text, key, ch))
    else:
      break
```

```python
#Diffie Hellman Key Exchange
from random import randint
# shared prime number and generator
q = 7
print("Prime Number:",q)
α = 3
print("Generator:",α)
# Alice's private key
XA = randint(1, q-1)
print("Private Key of A:",XA)
# Bob's private key
XB = randint(1, q-1)
print("Private Key of B:",XB)
# Alice's public key
YA = (α**XA) % q
print("Public Key of A:",YA)
# Bob's public key
YB = (α**XB) % q
print("Public Key of B:",YB)
# Shared secret key
k1 = (YB**XA) % q
k2 = (YA**XB) % q
if k1 == k2:
    print("Shared secret key:", k1)
else:
    print("Error: Shared secret keys do not match")
```
